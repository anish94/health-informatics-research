```python
# Importing required libraries and packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
```

```python
from datetime import datetime, timedelta
```

```python
#Reading the data file
DATA_PATH = "twitter_posts.csv"
data = pd.read_excel('twitter_posts.xls')
```

```python
data.head()
```

```python
data['Date (CST)'][0].date()
```

```python
data['Date (CST)'] = data['Date (CST)'].apply(lambda x: x.date())
```

```python
#d = datetime.strptime('2013-05-27', '%Y-%m-%d') # Monday
d = data['Date (CST)'][0]
t = timedelta((12 - d.weekday()) % 7)
(d + t).strftime('%Y-%m-%d')
```

```python
data['Date (CST)'] = data['Date (CST)'].apply(lambda x: (x + timedelta((12 - x.weekday()) % 7)).strftime('%Y-%m-%d'))
```
